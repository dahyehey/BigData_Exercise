```python
%%writefile src/mylib.py
def getKey(keyPath):
    d=dict()
    f=open(keyPath,'r')
    for line in f.readlines():
        row=line.split('=')
        row0=row[0]
        d[row0]=row[1].strip()
    return d
```

```python
%%writefile src/Practice3.py


import urllib
import os
import requests
import mylib
import xmltodict
import lxml
import lxml.etree

def saveFile(_fname,_data):
    fp=open(_fname,'w')
    fp.write(_data+"\n")
    
def doIt():
    keyPath=os.path.join(os.getcwd(), 'src', 'key.properties')
    key=mylib.getKey(keyPath)
    KEY=str(key['gokr'])
    
    serviceKey='serviceKey='+KEY
    page='page=1'
    countPerPage='countPerPage=1000'
    type='type=xml'
    params="&".join([serviceKey,page,countPerPage,type,])
    
    _url ='http://apis.data.go.kr/6500000/busNonstopService/busNonstopList?'
    url='&'.join([_url,params])

    r=requests.get(url).text
    print(r)
    
    _fname='src/P3_xml.txt'
    saveFile(_fname,r)


if __name__ == "__main__":
    doIt()
    
    
```
