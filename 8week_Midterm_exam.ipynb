## 문제 3: DataFrame을 사용해서 코로나 발생동향 분석

```python
import pyspark
import os
myConf = pyspark.SparkConf()
spark = pyspark.sql.SparkSession\
    .builder\
    .master("local")\
    .appName("myApp")\
    .config(conf=myConf)\
    .getOrCreate()
```

```python
myDf= spark.read.json(os.path.join("data","서울특별시 코로나19 자치구별 확진자 발생동향.json"))
```

## 2) (null제거하고) 종로, 마포, 용산, 합계 (종로, 마포, 용산) 10건 출력

```python
from pyspark.sql import functions as F
myDf2=myDf2.na.drop('any')
```

```python
from pyspark.sql.types import IntegerType

myDf2=myDf2.withColumn("jongno2", F.col('jongno').cast(IntegerType()))
myDf2=myDf2.withColumn("mapo2", F.col('mapo').cast(IntegerType()))
myDf2=myDf2.withColumn("yongsan2", F.col('yongsan').cast(IntegerType()))
```

```python
myDf2=myDf2.drop("jongno").drop("mapo").drop("yongsan")
```

## 문제 3-2의 답

```python
myDf2= myDf2.withColumn("tot", F.col("jongno2")+F.col("mapo2")+F.col("yongsan2"))
myDf2=myDf2.withColumnRenamed("jongno2","jongno")
myDf2=myDf2.withColumnRenamed("mapo2","mapo")
myDf2=myDf2.withColumnRenamed("yongsan2","yongsan")
myDf2.show(10)
```
